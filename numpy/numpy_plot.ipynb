```python
import numpy as np               # Import numpy 
import matplotlib.pyplot as plt  # Import matplotlib
# Inline plot figures
%matplotlib inline
```

```python
def gen_uncertain_data():
    x = np.linspace(0, 50, 200)
    y = np.sin(x/6*np.pi) + np.random.normal(0, 0.1, size=x.shape)
    error = np.random.normal(0.25, 0.1, size=y.shape)
    return x, y, error

a, b, e = gen_uncertain_data()
```

```python
fig, ax = plt.subplots()
plt.plot(a, b, '-', linewidth=2.0, color='blue')
plt.plot(a, np.max(b)+e, linewidth=1.5, color='green')
plt.plot(a, np.min(b)-e, linewidth=1.5, color='green')
plt.xlim([0, 50])
ax.fill_between(a, np.max(b)+e, np.min(b)-e, color=[0.75, 0.85, 0.85])
fig.set_facecolor([0.9, 0.9, 0.9])
ax.set_facecolor([0.6, 0.6, 0.6])

plt.grid()
```
